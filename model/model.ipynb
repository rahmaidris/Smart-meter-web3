## format the dataset :

```python



hourofday = []
minuteofhour = []
dayofweek = []
dayofmonth = []
monthofyear = []
year = []
consumption = []

# Loop through each row of the DataFrame
for index, row in df.iterrows():
    # Extract the date string and convert it to a datetime object
    date_str = row['date']
    dt = datetime.strptime(date_str, '%d/%m/%Y')


    # Loop through each time slot in the row and extract the required values
    for time_slot, cons in row.items():
        if time_slot in ['Customer', 'Generator Capacity', 'Postcode', 'Consumption Category', 'date', 'Row Quality']:
            continue  # Skip non-data columns

        # Extract the hour and minute values from the time slot string
        hour, minute = map(int, time_slot.split(':'))

        # Append the extracted values to their respective lists
        hourofday.append(hour)
        minuteofhour.append(minute)
        dayofweek.append(dt.weekday())
        dayofmonth.append(dt.day)
        monthofyear.append(dt.month)
        year.append(dt.year)
        consumption.append(float(cons))

# Create the output DataFrame with the extracted values
output_df = pd.DataFrame({
    'hourofday': hourofday,
    'minuteofhour': minuteofhour,
    'dayofweek': dayofweek,
    'dayofmonth': dayofmonth,
    'monthofyear': monthofyear,
    'year': year,
    'consumption': consumption
})

# Print the output DataFrame
print(output_df.head())
```

```python
output_df.shape[0]/df.shape[0]
```

```python
# Add metadata columns to the output DataFrame using the assign() method
output_df = output_df.assign(
    Customer=df['Customer'].repeat(48).values,
    Generator_Capacity=df['Generator Capacity'].repeat(48).values,
    Postcode=df['Postcode'].repeat(48).values,
    Consumption_Category=df['Consumption Category'].repeat(48).values,
    Row_Quality=df['Row Quality'].repeat(48).values
)

# Print the modified DataFrame
print(output_df.head())
```

```python
Enc = OrdinalEncoder()
output_df["year"] = Enc.fit_transform(output_df[["year"]]).astype('int64')
output_df[["year"]].value_counts()
```

```python
Enc_1 = OrdinalEncoder()
output_df["Customer"] = Enc_1.fit_transform(output_df[["Customer"]]).astype('int64')
output_df[["Customer"]].value_counts()
```

## add the anomaly column :

```python
# Group the data by customer and calculate the 90th percentile
q3 = output_df.groupby('Customer')['consumption'].quantile(0.9)
```

```python
q3
```

```python
# Define a function to set the anomaly flag
def set_anomaly(row):
    if row['consumption'] > q3[row['Customer']]:
        return 1
    else:
        return 0

# Apply the function to each row
output_df['anomaly'] = output_df.apply(set_anomaly, axis=1)
```

```python
# Count the number of occurrences of each value
counts = output_df['anomaly'].value_counts()

# Calculate the percentage of 1s and 0s
percentage_1 = counts[1] / len(output_df) * 100
percentage_0 = counts[0] / len(output_df) * 100
print(percentage_0,percentage_1)
```

## add profil column :

```python
from sklearn.cluster import KMeans


# Step 1: Group customers by postcode and calculate mean consumption for each customer
customer_mean_consumption = output_df.groupby(['Postcode', 'Customer'])['consumption'].mean().reset_index()

# Step 2: Calculate the mean consumption for each group of customers with the same postcode
postcode_mean_consumption = customer_mean_consumption.groupby('Postcode')['consumption'].mean().reset_index()

# Step 3: Use K-means to cluster postcode groups into 3 profiles based on their mean consumption
kmeans = KMeans(n_clusters=3, random_state=42)
postcode_mean_consumption['Profile'] = kmeans.fit_predict(postcode_mean_consumption[['consumption']])

# Step 4: Map the profile labels back to each customer in the main DataFrame
output_df = output_df.merge(postcode_mean_consumption[['Postcode', 'Profile']], on='Postcode', how='left')

# Step 5: Rename the 'Profile' column to 'Profile_Label'
output_df.rename(columns={'Profile': 'Profile_Label'}, inplace=True)

# Display the updated DataFrame
print(output_df.head())
```

```python
def evaluation(y_true,y_pred,y_pred_proba):

    con_matrix = confusion_matrix(y_true, y_pred)
    tn,fp,fn,tp =con_matrix.flat

    classes=[0,1]
    fig, ax = plt.subplots()
    sns.heatmap(pd.DataFrame(con_matrix), annot=True,fmt='g', cmap="Blues")
    ticks = np.arange(len(classes))
    plt.xticks(ticks, classes)
    plt.yticks(ticks, classes)
    plt.title('Confusion matrix')
    plt.ylabel('True labels')
    plt.xlabel('Predicted labels')

#==================================================================
    Sensitivity = round(tp/(tp+fn),3)
    Specificity = round(tn/(tn+fp),3)
    G_mean=round(np.sqrt(Sensitivity*Specificity),3)
    print("\nSensitivity = Recall", Sensitivity,"\nSpecificity ",Specificity,"\nG_mean ",G_mean)

    fp_rate = round(fp/(fp+tn),3)
    print("\n False positive rate", fp_rate )
    print("\n True positive Rate ",Sensitivity)
    F2 = round(fbeta_score(y_true, y_pred, average='binary', beta=2),3)
    print("\n F2-score",F2)
      #===================================================================

    print("\n\n\n The classification_report")
    print(classification_report(y_true,y_pred, labels=[0,1]))

    #==================================================================

    P =round(precision_score(y_true, y_pred, average='weighted'),3)
    R = round(recall_score(y_true, y_pred, average='weighted'),3)
    F1 = round(f1_score(y_true, y_pred, average='weighted'),3)
    roc_auc = []
    if (np.array_equal(y_pred_proba,np.array([]))) != True:
        print("\n ROC curve + AUC score")
        fpr, tpr, _ = roc_curve(y_true, y_pred_proba)
        roc_auc = round(roc_auc_score(y_true, y_pred_proba),3)
        plt.plot(fpr,tpr,label="AUC="+str(roc_auc),marker='.')
        plt.xlabel('FPR')
        plt.ylabel('TPR')
        plt.legend()
        plt.show()

    #===================================================================

    scores=dict()
    scores['Sensitivity']=Sensitivity
    scores['Specificity']=Specificity
    scores['G_mean']=G_mean
    scores['False_positive_rate']=fp_rate
    scores['Precision']=P
    scores['Recall']=R
    scores['F1-score']=F1
    scores['F2-score']=F2
    scores['roc_auc_score']=roc_auc

    return scores
```

```python
start = time.time()
model_1 = LogisticRegression(class_weight='balanced',max_iter=5000)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

metric_F1 = cross_val_score(model_1, x_train, y_train, scoring='f1', cv=cv)
print('Mean F1_score: %.3f' % np.mean(metric_F1) )
metric_precision = cross_val_score(model_1, x_train, y_train, scoring='precision', cv=cv)
print ('Mean precision: %.3f'%np.mean(metric_precision))
metric_recall = cross_val_score(model_1, x_train, y_train, scoring='recall', cv=cv)
print ('Mean Recall: %.3f'%np.mean(metric_recall))
metric_accuracy = cross_val_score(model_1, x_train, y_train, scoring='accuracy', cv=cv)
print ('Mean accuracy: %.3f'%np.mean(metric_accuracy))
print("Logistic Regression with original data is done in %0.2fs" % (time.time() - start))
```

```python
start = time.time()
model_2 = LogisticRegression( class_weight='balanced',max_iter=5000)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scale = StandardScaler()
pipeline_1 = Pipeline(steps=[('scale', scale), ('model', model_2)])

metric_F1 = cross_val_score(pipeline_1, x_train, y_train, scoring='f1', cv=cv)
print('Mean F1_score: %.3f' % np.mean(metric_F1))

metric_precision = cross_val_score(model_1, x_train, y_train, scoring='precision', cv=cv)
print ('Mean precision: %.3f'%np.mean(metric_precision))
metric_recall = cross_val_score(model_1, x_train, y_train, scoring='recall', cv=cv)
print ('Mean Recall: %.3f'%np.mean(metric_recall))
metric_accuracy = cross_val_score(model_1, x_train, y_train, scoring='accuracy', cv=cv)
print ('Mean accuracy: %.3f'%np.mean(metric_accuracy))
print("Logistic Regression with resmapling and rescaling method is done in %0.2fs" % (time.time() - start))
```

```python
#Hyperparameter Tuning for LogisticRegression
#The main hyperparameters we will tune in logistic regression are solver, penalty,
#max_iter, C (regularization strength), tol, fit_intercept, intercept_scaling, class_weight, random_state, multi_class, verbose, warm_start, and l1_ratio
start = time.time()
LR_model = LogisticRegression(max_iter=5000,multi_class='ovr',verbose=True)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=0)
balance = [{0:10,1:1}, {0:1,1:1}, {0:1,1:10},'balanced']
param_grid = dict(model__class_weight=balance)
param_grid['model__C'] = [0.01, 0.1, 1, 10]
param_grid['model__solver'] = ['lbfgs', 'liblinear','saga']
param_grid['model__penalty'] = ['l1', 'l2']
scaling =StandardScaler()
pipe = Pipeline(steps=[('scaler', scaling), ('model', LR_model)])
search = GridSearchCV(pipe, param_grid ,scoring='f1', cv=cv)
best_model = search.fit(x_train, y_train)
end = time.time()

print('Best F1-Score: %s' % best_model.best_score_)
print('Best Hyperparameters: %s' % 0.923)
print('Time consumed is: ',end - start)
```
